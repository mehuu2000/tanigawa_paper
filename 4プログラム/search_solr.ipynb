# 目的
* Pysolrを用いたApache Solrでの全文検索
* curlコマンドを用いたApache Solrでの全文検索

## <font color="red">要実行</font>

```python
# モジュールの読み込み
import pysolr
import generate_query   # 自作関数

# 接続文字列
solr_url = "http://localhost:8983/solr/tanigawa_jalc"

# pysolrのクライアントの初期化
solr = pysolr.Solr(solr_url,timeout=10)
```

## Pysolrでの検索

```python
# 検索文字列
query = "[前田 98] 前田晴美，梶原史雄，足立秀和，沢田篤史，武田英明，西田豊明：弱い情報共有を用いたコミュニティの情報共有システム，\
システム制御情報学会論文誌, Vol. 11, No. 10, pp. 568--575, 1998."

# エスケープ処理
escaped_query = generate_query.escape_solr_query(query)

# 検索オプションの設定
search_options = {
    'q.op':'OR',
    'fl':'doi,creator,title,journal,issued,volume_issue,page_range,score',
    'df':'jalcdata',
    'row':10
}

# 検索
try:
    results = solr.search(escaped_query,**search_options)   # solrで検索・結果の取得
    print(f"Total results found: {results.hits}")
    print(f"Number of results retrieved: {len(results)}") # 実際に取得された件数
    for result in results:
        print(result)
except pysolr.SolrError as e:
    print("エラー発生")       
```
