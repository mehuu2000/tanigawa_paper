## 特定のカラム（フィールド・列）のみを表示

* ```find()```メソッドの<mark>2つ目の引数</mark>に表示させたいカラムを指定

| （各属性に対する）値 | 結果 |
| :---: | --- |
| 0 | 非表示 |
| 1 | 表示 |
* 0だけを指定すれば、それ以外の属性は表示される（逆も然り）
* 引数に何も入力しなければ全てのカラムが表示される
<details>
<summary>（例）</summary>

```find({},{"_id":0,"doi":1,"title_list":1})```

上記のコードだと

| 属性 | 表示：非表示 |
| ---- | ---- |
| _id | 非表示 |
| doi | 表示 |
| title_list | 表示 |

</details>

````python
#mongodbへの接続
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

for x in mycol.find({},{"_id":0,"doi":1,"title_list":1}):
    print(x)
    break
````

## 検索
* 色々と検索をしてみる

### 抽出条件を指定した検索
* _idへの条件指定の場合は特殊
<details>
<summary>説明</summary>

* _id属性はMongoDB側から自動的に割り当てられる
* _idでの検索は次のように

~~~python
from bson import ObjectId
object_id = ObjectId(番号)
result = mycol.find_one({"_id": object_id})
~~~
</details>

* それ以外は```{属性名: 値}```の形で検索可能

````python
#_idフィールドでの検索
from bson import ObjectId

myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

object_id = ObjectId("67503f7c451fb595f3d8d401")
result = mycol.find_one({"_id": object_id},{"_id":0,"doi":1,"title_list":1})
print(result)
````

````python
#それ以外のフィールドでの検索

myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

result = []

for x in mycol.find({"volume": "20"},{"_id":0,"doi":1,"title_list":1}):
    result.append(x)

print(f"{len(result)}件ヒットしました。")
print("="*20)
print(result[0])
````

#### 部分一致による検索 **regex**
```
find("属性名":{"$regex":"条件文字列"})
```

````python
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

for x in mycol.find({"title_list.title":{"$regex":"書誌情報抽出"}}):
    print(x)
    break
                                         
````

複数のワードによる部分一致を行いたい場合
```
find("属性名":{"$regex":"単語1|単語2"})
```

````python
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

ans = []
for x in mycol.find({"title_list.title":{"$regex":"自然言語処理|構文解析"}},{"_id":0,"title_list.title":1}):
    ans.append(x)

print(f"検索件数：{len(ans)}件")
for i in range(10):
    print(ans[i])    
````

#### 以上($gte)
```
find({"属性名":{"$gte":数値}})
```
* 当該データは数値も文字列として登録されているので型変換を行う必要がある
```
find({"$expr":{演算子:[{"$toInt":"$属性名"},数値]}})
```
* $expr : 式の評価
* $toInt : 文字列をint型に

````python
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

ans = []

for x in mycol.find({"$expr": {"$gte": [{"$toInt": "$publication_date.publication_year"}, 2024]}}):
    ans.append(x)

print(f"検索件数：{len(ans)}件")    
````

#### 超過(&gt)
```
find({"属性名":{"$gt":数値}})
```

````python
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

result = []

for x in mycol.find({"$expr":{"$gt":[{"$toInt":"$publication_date.publication_year"},2023]}}):
    result.append(x)

print(f"検索件数：{len(result)}件")
for ans in result:
    print(ans)
    break    
````

#### 以下($lte)

find({"属性名":{"$lte":数値}})

### 複数条件

#### AND条件
* 全ての条件を満たすものが検索される
```python
find({"$and":[{条件1},{条件2},・・・]})
```

````python
#2020年以降に出版された論文のうち、「自然言語処理」を含むものを検出
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

result = []

for x in mycol.find({"$and":[{"title_list.title":{"$regex":"自然言語処理"}},{"$expr":{"$gte":[{"$toInt":"$publication_date.publication_year"},2020]}}]},\
                    {"_id":0,"title_list.title":1,"publication_date.publication_year":1}):
    result.append(x)

print(f"検索件数：{len(result)}件")
print(result[0])
print(result[1])    
````

#### OR条件
* 条件を一つでも満たせば検索の対象となる
```python
find({"$or":[{条件1},{条件2},・・・]})
```

````python
#「自然言語処理」もしくは「形態素解析」を含むものを検索
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]

result = []

for x in mycol.find({"$or":[{"title_list.title":{"$regex":"自然言語処理"}},{"title_list.title":{"$regex":"形態素解析"}}]},\
                    {"_id":0,"title_list.title":1}):
    result.append(x)

print(f"検索件数：{len(result)}件")
for i in range(5):
    print(result[i])
````
