# jalcdataフィールドで出現した単語について

```python
import requests
import json
import pprint

# --- 設定情報 ---
SOLR_HOST = 'http://localhost:8983/solr' # Solrのホスト名とポート
CORE_NAME = 'jalc'              # ターゲットとするコア名またはコレクション名
FIELD_NAME = 'jalcdata'                   # 最も出現した単語を知りたいフィールド名 (例: 日本語テキストフィールド)
LIMIT = 100                               # 取得したい上位単語の数
# -----------------

SOLR_URL = f"{SOLR_HOST}/{CORE_NAME}/terms"

# Solr Terms Component に送るパラメータ
params = {
    'terms': 'true',          # Terms Componentを有効にする
    'terms.fl': FIELD_NAME,   # どのフィールドを分析するか
    'terms.sort': 'count',    # 出現回数 (count) 順でソート (アルファベット順は index)
    'terms.limit': LIMIT,     # 取得する単語の最大数
    'wt': 'json'              # レスポンス形式をJSONに指定
}

print(f"リクエストURL: {SOLR_URL}")
print(f"パラメータ: {params}")

try:
    # HTTP GETリクエストを送信
    response = requests.get(SOLR_URL, params=params)
    response.raise_for_status() # ステータスコードが200番台以外の場合、例外を発生させる

    # JSONレスポンスをパース
    data = response.json()

    # --- 結果の整形と表示 ---
    terms_data = data.get('terms', {}).get(FIELD_NAME, [])
    
    if terms_data:
        print("\n--- 最も出現した単語 (上位2000件) ---")
        
        # Terms Component の結果は [単語1, カウント1, 単語2, カウント2, ...] の形式で返される
        # これを単語とカウントのペアに変換
        
        # 2つずつ要素を取り出す
        term_dict = {}
        for i in range(0, len(terms_data), 2):
            term = terms_data[i]
            count = terms_data[i+1]
            term_dict[term] = count
        for key,value in term_dict.items():
            print(f"{key} | {value}")  
          
    else:
        print("\n指定されたフィールドで単語が見つかりませんでした。フィールド名やインデックス設定を確認してください。")
        
except requests.exceptions.HTTPError as errh:
    print(f"\nHTTPエラーが発生しました: {errh}")
    print("SolrのURL、コア名、またはフィールド名を確認してください。")
except requests.exceptions.ConnectionError as errc:
    print(f"\n接続エラーが発生しました: {errc}")
    print("Solrサーバーが起動しているか、ホスト名とポート番号が正しいか確認してください。")
except requests.exceptions.RequestException as err:
    print(f"\nリクエスト中に予期せぬエラーが発生しました: {err}")
```
