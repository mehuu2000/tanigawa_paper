# 目的
* 人工知能学会論文誌 18巻に掲載されている論文から参考文献文字列を取得する

## <font color="Red">要実行</font>

```python
#モジュール及び接続文字列の読み込み
import pymongo
url_mongodb = "mongodb://localhost:27017/"

#mongodbへの接続
myclient = pymongo.MongoClient(url_mongodb)
mydb = myclient["jalc"]
mycol = mydb["restapi"]
```

## 取得プログラム

```python
#人工知能学会論文誌 18巻 ４号に掲載されている論文のDOI
url_list = [
    "https://doi.org/10.1527/tjsai.18.165",
    "https://doi.org/10.1527/tjsai.18.173",
    "https://doi.org/10.1527/tjsai.18.183",
    "https://doi.org/10.1527/tjsai.18.193",
    "https://doi.org/10.1527/tjsai.18.203",
    "https://doi.org/10.1527/tjsai.18.212",
    "https://doi.org/10.1527/tjsai.18.221"
]

print(len(url_list))
```

```python
#上記の論文から参考文献文字列を取得
projection = {
    "_id":0,
    "citation_list":1
}

query = {"url":{"$in":url_list}}

references = [] #参考文献文字列

for doc in mycol.find(query,projection):
    for reference in doc["citation_list"]:
        references.append(reference["original_text"])

print(len(references))
print(references[0])        
```

```python
#上記の参考文献文字列から和文を抽出
import re

#日本語を含むか判定
def has_japnese(text):
    return bool(re.search(r'[\u3040-\u309F\u30A0-\u30FF\uFF66-\uFF9F\u4E00-\u9FFF]', text))

#本処理
jp_references = []
for reference in references:
    if has_japnese(reference):
        jp_references.append(reference)

print(len(jp_references))
print(jp_references[-1])        
```

```python
# テキストファイルの作成
with open('jsai_reference.txt','x') as f:
    f.write('\n'.join(jp_references))
```
