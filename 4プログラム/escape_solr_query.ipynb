# apache solr の検索クエリに含まれる特殊文字をエスケープする
* solr内の特殊文字と半角スペースをエスケープしないと望む検索結果が得られなかった

```python
def escape_solr_query_robust(text):
    # Solrの特殊文字リスト (長い順、または '\' を最初にエスケープ) + 半角スペース
    # '\\' は必ず最初にエスケープしないと、他の文字をエスケープした後に発生する '\\' もエスケープされてしまう
    solr_special_chars_map = {
        '\\': r'\\', # バックスラッシュ自体をエスケープ
        '+': r'\+',
        '-': r'\-',
        '&': r'\&',
        '|': r'\|',
        '!': r'\!',
        '(': r'\(',
        ')': r'\)',
        '{': r'\{',
        '}': r'\}',
        '[': r'\[',
        ']': r'\]',
        '^': r'\^',
        '"': r'\"',
        '~': r'\~',
        '*': r'\*',
        '?': r'\?',
        ':': r'\:',
        '/': r'\/'
    }

    escaped_text = text
    # キーの順序が重要になるため、明示的に指定するか、一度に置き換える
    for char, escaped_char in solr_special_chars_map.items():
        escaped_text = escaped_text.replace(char, escaped_char)
    return escaped_text

query = input("クエリを入力 ->")
escaped_query = escape_solr_query_robust(query)
print(f"元クエリ->{query}")
print(f"エスケープ処理語のクエリ->{escaped_query}")
```
