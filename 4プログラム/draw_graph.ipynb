# 目的：関数の描画

## ライブラリの読み込み

```python
import matplotlib.pyplot as plt
import numpy as np
```

## グラフの描画

```python
# 定義域
x = np.linspace(0,2,1000)
# 関数
y1 = x/(1.2*((1-0.75)+0.75*2.0)+x)
y2 = x/(1.2*((1-0.75)+0.75*0.5)+x)
y3 = x/(1.2*((1-0.75)+0.75*1)+x)
# 表示範囲の設定
plt.xlim(0,1)
plt.ylim(0,1.2)
# メモリの表示
plt.xticks([0]) # x軸の値を消す
plt.yticks([0,1]) # y軸の値を消す

# 描画
plt.plot(x, y1, label=r'$dl>avdl$')
plt.plot(x,y2, label=r'$dl<avdl$')
plt.plot(x,y3,color='black',linestyle='--',label=r'dl=avdl')
plt.axhline(1, color = 'black', linestyle=':', alpha=0.5)
plt.xlabel('tf')
plt.ylabel(r'$\phi$')
plt.legend()
#plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
